<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• Imported  libraries</h4>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• 1 Selecting Image, and denote it by I :</h4>

```python
# I = cv2.imread("C:/Users/TRETEC/Desktop/watch_white_background.png")
I = cv2.imread("./watch_white_background.png")
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
plt.imshow(I, cmap = 'gray')
plt.show()
```

<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• 2 Resize the image to 100 x 100 : :</h4>

```python
I = cv2.resize(I, (100,100))
plt.imshow(I, cmap = 'gray')
plt.show()
```

<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• 3 Encode the image using Fourier transform :</h4>

$$ F(u,v) = \frac{1}{M \times N} \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} f(x,y) \cdot e^{-2\pi i \left( \frac{u \cdot x}{M} + \frac{v \cdot y}{N} \right)} $$


```python
def DFT2D(img):
    M, N = img.shape
    dft = np.zeros((M, N), dtype=np.complex64)
    for u in range(M):
        for v in range(N):
            sum_val = 0
            for x in range(M):
                for y in range(N):
                    sum_val += img[x, y] * np.exp(
                        -2j * np.pi * ((u * x) / M + (v * y) / N)
                    )
            dft[u, v] = sum_val / (M * N)
    return dft
```

```python
dft_I = DFT2D(I)
```

<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• 4 Recover the image using IDFT top left coefficients :</h4>

$$ f(x,y) =  \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} F(u,v) \cdot e^{2\pi i \left( \frac{u \cdot x}{M} + \frac{v \cdot y}{N} \right)} $$


```python
def inverse_dft(dft_coeffs, size):
    M, N = size, size
    recovered_image = np.zeros((M, N), dtype=np.complex128)

    for x in range(M):
        for y in range(N):
            sum_val = 0.0
            for u in range(M):
                for v in range(N):
                    e_term = np.exp(2j * np.pi * ((u * x) / M + (v * y) / N))
                    sum_val += dft_coeffs[u, v] * e_term
            recovered_image[x, y] = sum_val

    return np.abs(recovered_image)
```

```python
recovered_image_20x20 = inverse_dft(dft_I, 20)
recovered_image_50x50 = inverse_dft(dft_I, 50)
recovered_image_90x90 = inverse_dft(dft_I, 90)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 4, 1)
plt.imshow(recovered_image_20x20, cmap="gray")
plt.title("20x20 Coefficients")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(recovered_image_50x50, cmap="gray")
plt.title("50x50 Coefficients")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(recovered_image_90x90, cmap="gray")
plt.title("90x90 Coefficients")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(I, cmap="gray")
plt.title("Original image")
plt.axis("off")

plt.show()
```

<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• 5 Recover the image using IDFT bottom right coefficients :</h4>

```python
def inverse_dft_bottom_right(dft_coeffs, size):
    M, N = size, size
    recovered_image = np.zeros((M, N), dtype=np.complex128)

    for x in range(M):
        for y in range(N):
            sum_val = 0.0
            for u in range(len(dft_coeffs) - M, len(dft_coeffs)):
                for v in range(len(dft_coeffs) - M, len(dft_coeffs)):
                    sum_val += dft_coeffs[u, v] * np.exp(
                        2j * np.pi * ((u * x) / M + (v * y) / N)
                    )
            recovered_image[x, y] = sum_val

    return np.abs(recovered_image)
```

```python
recovered_image_20x20 = inverse_dft_bottom_right(dft_I, 20)
recovered_image_50x50 = inverse_dft_bottom_right(dft_I, 50)
recovered_image_90x90 = inverse_dft_bottom_right(dft_I, 90)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 4, 1)
plt.imshow(recovered_image_20x20, cmap="gray")
plt.title("20x20 Coefficients")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(recovered_image_50x50, cmap="gray")
plt.title("50x50 Coefficients")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(recovered_image_90x90, cmap="gray")
plt.title("90x90 Coefficients")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(I, cmap="gray")
plt.title("Original image")
plt.axis("off")

plt.show()
```

<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• 6 display amplitude image :</h4>

```python
def shift_dft(dft_coeffs):
    M, N = dft_coeffs.shape
    shifted_dft = np.zeros_like(dft_coeffs, dtype=np.complex128)

    for u in range(M):
        for v in range(N):
            shifted_u = (u + M // 2) % M
            shifted_v = (v + N // 2) % N
            shifted_dft[shifted_u, shifted_v] = dft_coeffs[u, v]

    return shifted_dft
```

```python
amplitude = np.sqrt(np.power(dft_I.real, 2) + np.power(dft_I.imag, 2))

plt.imshow(np.abs(shift_dft(amplitude)), cmap="gray")
plt.show()
```

<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• 7 Add gaussian noise to I, then denoise it using low pass filter :</h4>

```python
def normalized_inverse_dft(dft_coeffs, size):
    M, N = size, size
    recovered_image = np.zeros((M, N), dtype=np.complex128)

    for x in range(M):
        for y in range(N):
            sum_val = 0.0
            for u in range(M):
                for v in range(N):
                    e_term = np.exp(2j * np.pi * ((u * x) / M + (v * y) / N))
                    sum_val += dft_coeffs[u, v] * e_term
            recovered_image[x, y] = sum_val / (M * N)

    return np.real(recovered_image)


def ideal_low_pass_filter(dft_coeffs, cutoff_frequency):
    M, N = dft_coeffs.shape
    filtered_dft = np.zeros_like(dft_coeffs, dtype=np.complex128)

    for u in range(M):
        for v in range(N):
            distance = np.sqrt((u - M // 2) ** 2 + (v - N // 2) ** 2)
            if distance <= cutoff_frequency:
                filtered_dft[u, v] = dft_coeffs[u, v]
            else:
                filtered_dft[u, v] = 0

    return filtered_dft


def low_pass_filter(img, cutoff_frequency):
    # Compute the 2D DFT of the image
    dft_coeffs = DFT2D(img)

    # Shift the zero-frequency component to the center of the spectrum
    shifted_dft = shift_dft(dft_coeffs)

    # Apply the ideal low pass filter
    filtered_dft = ideal_low_pass_filter(shifted_dft, cutoff_frequency)

    # Shift the zero-frequency component back to the top-left corner of the spectrum
    shifted_filtered_dft = shift_dft(filtered_dft)

    # Compute the inverse 2D DFT of the filtered coefficients
    filtered_img = normalized_inverse_dft(shifted_filtered_dft, img.shape[0])

    return filtered_img
```

```python
noisy_I = np.array(I + np.random.normal(loc=0, scale=20, size=I.shape))
```

```python
filtered_img = low_pass_filter(noisy_I, 50)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
plt.imshow(noisy_I, cmap="gray")
plt.title("noisy image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(filtered_img, cmap="gray")
plt.title("low pass filtered noisy image")
plt.axis("off")
```

<h4 style="color: #95a5a6; font-family: 'Arial', sans-serif; padding: 20px;">• 8 Apply high pass filter on I :</h4>

```python
def ideal_high_pass_filter(dft_coeffs, cutoff_frequency):
    M, N = dft_coeffs.shape
    filtered_dft = np.zeros_like(dft_coeffs, dtype=np.complex128)

    for u in range(M):
        for v in range(N):
            distance = np.sqrt((u - M // 2) ** 2 + (v - N // 2) ** 2)
            if distance >= cutoff_frequency:
                filtered_dft[u, v] = dft_coeffs[u, v]
            else:
                filtered_dft[u, v] = 0

    return filtered_dft

def high_pass_filter(img, cutoff_frequency):
    # Compute the 2D DFT of the image
    dft_coeffs = DFT2D(img)

    # Shift the zero-frequency component to the center of the spectrum
    shifted_dft = shift_dft(dft_coeffs)

    # Apply the ideal high pass filter
    filtered_dft = ideal_high_pass_filter(shifted_dft, cutoff_frequency)

    # Shift the zero-frequency component back to the top-left corner of the spectrum
    shifted_filtered_dft = shift_dft(filtered_dft)

    # Compute the inverse 2D DFT of the filtered coefficients
    filtered_img = normalized_inverse_dft(shifted_filtered_dft, img.shape[0])

    return filtered_img
```

```python
filtered_img = high_pass_filter(noisy_I, 50)
```
